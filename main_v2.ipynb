```python
import pandas as pd
```

```python
df = pd.read_csv('Dataset_model.csv')

df = df[['novo_user_id', 'acertou', 'difficulty', 'discipline_id', 'institute_id', 'knowledge_area_id', 'modality_id', 'product_id', 'scholarity_id']]

df = df.fillna(0)

df = df.astype('int')

df
```

```python
X_train = df.drop('acertou', axis = 1)
y_train = df['acertou']

print(X_train.shape, y_train.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state = 42, n_jobs=-1)

rf.fit(X_train, y_train)

rf.score(X_train, y_train)
```

```python
from sklearn.metrics import f1_score

f1_score(y_train, rf.predict(X_train))
```

```python
df_test = pd.read_csv('Submit.csv', delimiter=';')

df_test = df_test[['novo_user_id', 'difficulty', 'discipline_id', 'institute_id', 'knowledge_area_id', 'modality_id', 'product_id', 'scholarity_id']]

df_test = df_test.fillna(0)

df_test = df_test.astype('int')

df_test
```

```python
y_pred = rf.predict(df_test)

y_pred

result = pd.Series(dtype='int', data=y_pred)
```
