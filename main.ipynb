## Prueba Real

```python
df_1 = pd.read_csv('Dataset_model.csv')

df_2 = pd.read_csv('Submit.csv', delimiter=';')

df = pd.concat([df_1, df_2], axis=0)

df = df.sort_values(by=['novo_user_id', 'row'])

df = df[['acertou', 'difficulty', 'discipline_id', 'institute_id', 'knowledge_area_id', 'modality_id', 'product_id', 'scholarity_id']]

df = df.fillna(0)

df = df.astype('int')

df
```

```python
result = pd.Series(dtype='int')
```

```python
result = pd.Series(dtype='int')

for i in range(0, df.shape[0], 101):
    batch = df.iloc[i:101+i, :]

    X_train = batch.drop('acertou', axis = 1).iloc[:100, :]
    y_train = batch['acertou'].iloc[:100]

    X_test = batch.drop('acertou', axis = 1).iloc[100, :]

    rf = RandomForestClassifier(n_estimators = 100, random_state = 42, max_depth=12)
    rf.fit(X_train, y_train)

    y_pred = rf.predict(X_test.values.reshape(1, -1))

    result = result.append(pd.Series(y_pred))

    print(i)


    
    
```

```python
df.shape[0]
```

```python
result
```

```python
result.to_csv('result.csv', index=False)
```
